```python
cap = cv2.VideoCapture('/content/video_2025-11-22_21-30-07.mp4')
```

```python
while True:
    success, img = cap.read()
    if not success or img is None:
        print("Failed to read frame or video stream ended.")
        break
    cv2_imshow(img)
    if cv2.waitKey(1) & 0xFF == ord ('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
